```python
import json, re
# Load data
filename = "strings"
with open(f"../dj-visualizations/data/{filename}.json") as f:
    strings = json.load(f)

strings = list(filter(lambda x: re.search(r"\s", x), strings))
```

```python
import utils
from openai import OpenAI

client = OpenAI(api_key=utils.get_OPENAI_API_KEY_DJ())

def classify_pattern(prompt):
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "system", 
                "content": 
                """
                Read this thoroughly: https://github.com/corralm/awesome-prompting#prompt-pattern-catalog

                The prompt categories are:
                1. Input Semantics
                2. Output Customization
                3. Error Identification
                4. Prompt Improvement
                5. Interaction
                6. Context Control
                7. None of the above
                """
            },
            {
                "role": "user", 
                "content": f"""
                \"{prompt}\"
                
                Please categorize the given string into one of the categories above. 
                Respond with the number of the category. Only state the number.
                """
            }
        ],
        model="gpt-3.5-turbo",
        max_tokens=1,
        temperature=0,
    )

    return chat_completion.choices[0].message.content
```
